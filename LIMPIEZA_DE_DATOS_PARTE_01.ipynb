```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.encode("ascii", errors="ignore")
    .str.decode("ascii")
)
print(df.columns.tolist())
```

```python
df.head()
```

```python
# Nombre: capitalización consistente
df["nombre"] = df["nombre"].str.strip().str.title()

# Diagnóstico: mayúsculas y sin espacios
df["diagnostico"] = df["diagnostico"].str.strip().str.upper()

# Sexo: unificar categorías antes de convertir
mapa_sexo = {
    "m": "M", "masculino": "M",
    "f": "F", "femenino": "F",
    "999": np.nan
}
df["sexo"] = df["sexo"].str.strip().str.lower().map(mapa_sexo)
print(df["sexo"].value_counts(dropna=False))
```

```python
df.head()
```

```python
# peso_kg: reemplazar "N/A" string antes de convertir
df["peso_kg"] = df["peso_kg"].replace("N/A", np.nan)
df["peso_kg"] = pd.to_numeric(df["peso_kg"], errors="coerce")

# costo_usd: eliminar $ y comas
df["costo_usd"] = (
    df["costo_usd"]
    .str.replace("$", "", regex=False)
    .str.replace(",", "", regex=False)
    .str.strip()
)
df["costo_usd"] = pd.to_numeric(df["costo_usd"], errors="coerce")

print(df[["peso_kg", "costo_usd"]].dtypes)
```

```python
df.head()
```

```python
df["fecha_ingreso"] = pd.to_datetime(df["fecha_ingreso"], dayfirst=False, infer_datetime_format=True, errors="coerce")
df["fecha_alta"]    = pd.to_datetime(df["fecha_alta"],    dayfirst=False, infer_datetime_format=True, errors="coerce")

# Verificar cuántas fechas no se pudieron parsear
print(df["fecha_ingreso"].isnull().sum())
print(df["fecha_alta"].isnull().sum())
```

```python
df.head()
```

```python
# Edad: rango biológicamente plausible
mask_edad = (df["edad"] < 0) | (df["edad"] > 120)
print(f"Edades fuera de rango: {mask_edad.sum()}")
df.loc[mask_edad, "edad"] = np.nan

# Peso: rango plausible para humanos adultos
mask_peso = (df["peso_kg"] < 2) | (df["peso_kg"] > 300)
df.loc[mask_peso, "peso_kg"] = np.nan
```

```python
# Alta no puede ser anterior al ingreso
mask_fechas = df["fecha_alta"] < df["fecha_ingreso"]
print(f"Inconsistencias fecha ingreso/alta: {mask_fechas.sum()}")
df.loc[mask_fechas, "fecha_alta"] = np.nan
```

```python
print(df.isnull().sum())

# Edad y peso: imputar con mediana (robusta a atípicos)
df["edad"]    = df["edad"].fillna(df["edad"].median())
df["peso_kg"] = df["peso_kg"].fillna(df["peso_kg"].median())

# n_medicamentos: imputar con moda
moda_med = df["n_medicamentos"].mode()[0]
df["n_medicamentos"] = df["n_medicamentos"].fillna(moda_med)

# sexo: dejar como NaN si no es recuperable
# fecha_alta: dejar como NaN, es un dato clínico no imputable

print(df.isnull().sum())
```

```python
df["sexo"]           = df["sexo"].astype("category")
df["diagnostico"]    = df["diagnostico"].astype("category")
df["n_medicamentos"] = df["n_medicamentos"].astype(int)
df["id_paciente"]    = df["id_paciente"].astype(int)

print(df.dtypes)
```

```python
# Diccionario de listas: destino -> [variantes]
mapa_sexo_listas = {
    "M": ["m", "masculino", "male", "hombre"],
    "F": ["f", "femenino", "female", "mujer"],
    np.nan: ["999", "ns/nr", "desconocido", ""]
}

# Invertir: cada variante apunta a su destino
mapa_invertido = {
    variante: destino
    for destino, variantes in mapa_sexo_listas.items()
    for variante in variantes
}

```
